```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import numpy as np
import matplotlib.pyplot as plt
import cv2  

# Load the saved model
plant_model = load_model('models/planthealthmodel.h5')

# Load and preprocess a new image 
img = cv2.imread('1.jpg')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
img_resized = cv2.resize(img, (256, 256))  # Resize image to the size the model expects
img_normalized = img_resized / 255.0  # Normalize image values

#Expand dimensions to match the input shape
img_input = np.expand_dims(img_normalized, axis=0)  

#   Make predictions with the loaded model
plant_prediction = plant_model.predict(img_input)

if plant_prediction > 0.5:
    print('Plant is sick')
else:
    print('Plant is healthy')
```
